### 8) Жаккардово сходство текстов
```markdown
Сходство текстов часто считают по коэффициенту Жаккара на множествах слов. Требуется реализовать функцию jaccard(a: str, b: str, *, min_len=1) -> float, которая нормализует оба текста, строит множества слов и возвращает |A∩B|/|A∪B| (если объединение пусто — 0.0). Нормализация в виде обработки регистра/диакритики позволяет получить более высокую точность решения (плюсом будет посмотреть и другие техники нормализации слов).

Примеры:
a="Café au lait", b="cafe au Lait!" → 1.0
a="кот котик", b="Котёнок", min_len=3 → 0.0 (если не делать нормализацию)
